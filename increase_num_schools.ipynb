```python
# soup.find('div', class_='organic-schools desktop-inner-wrapper')
schools = soup.find_all('div', class_='school')
```

```python
print(BeautifulSoup(str(schools)).prettify())
```

```python
school_name = BeautifulSoup(str(schools)).find_all('p', class_='name')
school_address = BeautifulSoup(str(schools)).find_all('p', class_='address')
```

```python
names_clean = re.sub(r'<.*?>', '', str(school_name)).split(', ')
addresses_clean = re.sub(r'<.*?>', '', str(school_address)).split(', \n')

for i in range(len(addresses_clean)):
    addresses_clean[i] = addresses_clean[i].replace('\n', '').replace('  ', '').replace(']', '').replace('[', '')
```

```python
schools_dict = {}

for i in range(len(names_clean)):
    # print(names_clean[i], ' - ', addresses_clean[i])
    try:
        schools_dict[names_clean[i]] = addresses_clean[i]
    except:
        pass
```
