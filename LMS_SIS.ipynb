```python
url = 'https://stats.next.moodle.org/sites/index.php?country=US'

req = urllib.request.Request(url)
response = urllib.request.urlopen(req).read()
soup = BeautifulSoup(response)
```

```python
print(soup.prettify())
```

```python
list_items = BeautifulSoup(str(soup.find_all('div', class_='boxwidthwide boxaligncenter')))
list_items = list_items.find_all('li')
```

```python
BeautifulSoup(str(list_items[0])).find(href=True)['href']
```

```python
moodle_df = pd.DataFrame()

for i in range(len(list_items)):
    try:
        school = BeautifulSoup(str(list_items[i])).find(href=True)

        school_name = school.text
        school_site = school['href']
        moodle_df = moodle_df.append({'School Name':school_name, 'Moodle Site':school_site}, ignore_index=True)
    except:
        school_name = BeautifulSoup(str(list_items[i])).text
        moodle_df = moodle_df.append({'School Name':school_name, 'Moodle Site':''}, ignore_index=True)
    
```
