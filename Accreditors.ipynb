```python
def add_space(text):
    new_text = ''
    for i, letter in enumerate(text):
        if i and letter.isupper() and (text[i-1].islower() or text[i-1].isdigit()):
            new_text += ' '
        new_text += letter
    return new_text
```

```python
accet_df['Address'] = ''
accet_detail_df = pd.DataFrame()

for i in range(0, 5): #range(len(accet_df))
    url = accet_df['Additional Info'][i]
    req = urllib.request.Request(url)
    response = urllib.request.urlopen(req).read()
    soup = BeautifulSoup(response)

    school_val = {}
    titles = BeautifulSoup(str(soup.find_all('dl')[0])).find_all('dt') + BeautifulSoup(str(soup.find_all('dl')[1])).find_all('dt')
    values = BeautifulSoup(str(soup.find_all('dl')[0])).find_all('dd') + BeautifulSoup(str(soup.find_all('dl')[1])).find_all('dd')

    for t in range(len(titles)):
        try:
            school_val[titles[t].text] = values[t].text.replace('\n', '')
        except:
            continue

    # All Campuses
    all_addresses = soup.find('div', class_='span12')
    each_address = all_addresses.find('dd').text.split('\n\n')
    each_address.pop(-1)
    name_address_dict = {}

    for a in range(len(each_address)):
        if a%2 == 0 or a == 0:
            name_address_dict[each_address[a].replace('\n', '')] = add_space(each_address[a + 1])
        else:
            continue

    # Programs and their clock hours
    clock_hour_dict = {}
    programs = BeautifulSoup(str(soup.find_all('table'))).find_all('tr')

    for p in range(len(programs)):
        program_hours = BeautifulSoup(str(programs[p])).text.replace('\n', '').split('clock hours: ')
        clock_hour_dict[program_hours[0]] = program_hours[1]

    # Put into new dataframe
    school_val['Name'] = accet_df['Name'][i]
    school_val['Phone'] = accet_df['Phone'][i]
    school_val['City'] = accet_df['City'][i]
    school_val['ACCET ID'] = accet_df['ACCET ID'][i]
    school_val['Clock Hour Programs'] = clock_hour_dict
    school_val['Address'] = add_space(school_val['Address'])
    school_val['All Addresses'] = name_address_dict
    accet_detail_df = accet_detail_df.append(school_val, ignore_index=True)


accet_detail_df = accet_detail_df.reindex(['ACCET ID', 'Name', 'Website', 'Phone', 'Address', 'City', 'Member Type', 'Initial Accreditation', 'Accreditation Expires', 'Clock Hour Programs', 'All Addresses'], axis=1)
```
